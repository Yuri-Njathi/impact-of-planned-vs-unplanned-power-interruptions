Merge population data with counties 

```python
import pandas as pd

def get_merged_df(df1_path,df2_path,common="CODE",new_name="kenya_census_data.csv",how="outer",path=True):
    if path:
        df1 = pd.read_csv(df1_path)
        df2 = pd.read_csv(df2_path)
    
    merged = pd.merge(df1, df2, on=common, how=how) 
    print('Visualize Dataframe : \n',merged.head())
    print('\n Dataframe shape= ',merged.shape)
    merged.to_csv(new_name, index=False)
    return merged
```

```python
#merge population data with 
merged = get_merged_df(df1_path="/kaggle/input/counties-data/Kenya_counties_census_data_v2.csv",df2_path="/kaggle/input/counties-data/kenya_counties_v3.csv",common="CODE",new_name="kenya_census_data.csv",how="outer")
```

```python
#remove commas
oi["Conventional Households"] = oi["Conventional Households"].str.replace(",", "").astype(int)
```

```python
oi
```

```python
# conventional households using mains electricity
oi["CHME"] = round(oi["Conventional Households"]*oi["Mains Electricity"]/100,0)
#people per household
oi["PPH"] = round(oi["TOTAL_POP"]/oi['Conventional Households'])
```

```python
oi["CHME"] = oi["CHME"].astype(int)
```

```python
oi.to_csv("/kaggle/working/kenya_census_data.csv", index=False)
```

```python
final_census_df = oi[['CODE','Conventional Households','Mains Electricity','UNIT_AREA','UNIT_PERIM','COUNT','COUNTY_NAM','TOTAL_POP','POP_DENSIT','Latitude','Longitude','CHME','PPH']].copy()
```
