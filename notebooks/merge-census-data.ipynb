Merge population data with counties 

```python
import pandas as pd

df1 = pd.read_csv("/kaggle/input/counties-data/Kenya_counties_census_data_v2.csv")
df2 = pd.read_csv("/kaggle/input/counties-data/kenya_counties_v2.csv")

merged = pd.merge(df1, df2, on="CODE", how="outer")  
merged.to_csv("kenya_census_data.csv", index=False)
```
